Preparing data, tools and dependencies

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plot
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Dataset Loading and Analysis

```python
iris = load_iris()
dataset = pd.DataFrame(iris.data, columns = iris.feature_names)
dataset['flower_type'] = iris.target
dataset['flower_name'] = [iris.target_names[i] for i in iris.target]
print(dataset)
flowers = {1 : 'setosa', 2 : 'versicolor', 3 : 'virginica'}
colors = {'setosa':'red', 'versicolor':'green', 'virginica':'blue'}

# Visualise and Analyze Classification according Petal Stats of flower
plot.scatter(dataset['petal length (cm)'],dataset['petal width (cm)'],c=dataset['flower_name'].map(colors))
plot.xlabel("Petal Length (cm)")
plot.ylabel("Petal Width (cm)")
plot.show()

# Visualise and Analyze Classification according Sepal Stats of flower
plot.scatter(dataset['sepal length (cm)'],dataset['sepal width (cm)'],c=dataset['flower_name'].map(colors))
plot.xlabel("Sepal Length (cm)")
plot.ylabel("Sepal Width (cm)")
plot.show()

# plot.scatter(dataset['petal length (cm)'],dataset['sepal width (cm)'],c=dataset['flower_name'].map(colors))
# plot.xlabel("Petal Length (cm)")
# plot.ylabel("Sepal Width (cm)")
# plot.show()

# plot.scatter(dataset['petal length (cm)'],dataset['sepal length (cm)'],c=dataset['flower_name'].map(colors))
# plot.xlabel("Petal Length (cm)")
# plot.ylabel("Sepal Length (cm)")
# plot.show()

# Best Seperation of classes : Sepal Length vs Petal Length
plot.scatter(dataset['sepal length (cm)'],dataset['petal length (cm)'],c=dataset['flower_name'].map(colors))
plot.ylabel("Petal Length (cm)")
plot.xlabel("Sepal Length (cm)")
plot.show()

# plot.scatter(dataset['petal width (cm)'],dataset['sepal width (cm)'],c=dataset['flower_name'].map(colors))
# plot.xlabel("Petal Width (cm)")
# plot.ylabel("Sepal Width (cm)")
# plot.show()

# plot.scatter(dataset['petal width (cm)'],dataset['sepal length (cm)'],c=dataset['flower_name'].map(colors))
# plot.xlabel("Petal Width (cm)")
# plot.ylabel("Sepal Length (cm)")
# plot.show()

# plot.scatter(dataset['sepal length (cm)'],dataset['petal width (cm)'],c=dataset['flower_name'].map(colors))
# plot.ylabel("Petal Width (cm)")
# plot.xlabel("Sepal Length (cm)")
# plot.show()
```

Training ML Model : Logistic (Softmax) Regression for 3-class classification (with feature selection, elimination)

```python
# Using [Petal Width, Petal Length] for Classification
x = dataset[['petal width (cm)', 'petal length (cm)']]
y = dataset['flower_type']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print(f"[Petal Width, Petal Length] Accuracy : {accuracy_score(y_test,y_pred):.2f}")

# Using [Sepal Width, Sepal Length] for Classification
x = dataset[['sepal width (cm)', 'sepal length (cm)']]
y = dataset['flower_type']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print(f"[Sepal Width, Sepal Length] Accuracy : {accuracy_score(y_test,y_pred):.2f}")

# Using [Sepal Length, Petal Length] for Classification
x = dataset[['sepal length (cm)', 'petal length (cm)']]
y = dataset['flower_type']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print(f"[Sepal Length, Petal Length] Accuracy : {accuracy_score(y_test,y_pred):.2f}")

# Using [Petal Length] for Classification
x = dataset[['petal length (cm)']]
y = dataset['flower_type']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print(f"[Petal Length] Accuracy : {accuracy_score(y_test,y_pred):.2f}")
```
