```python
# Linear Regression Using Normal Equation
def LinearRegression (X, y) :

  X = np.array(X)
  y = np.array(y)

  if X.ndim == 1:  # If X is a 1D vector, reshape it to (m,1)
        X = X.reshape(-1, 1)
  X = np.c_[np.ones((len(X),1)) * 1, X] # bias = 1

  X_t = np.transpose(X)
  theta = np.linalg.inv(X_t @ X) @ (X_t @ y) # Theta = (X_transpose * X)^(-1) * X_transpose * y

  return LinearRegressionModel(theta)

class LinearRegressionModel :

  def __init__ (self, theta) :
    self.parameters = theta

  def predict (self, X) : # y = X * Theta

    X = np.array(X)
    if X.ndim == 1:
      X = X.reshape(-1, 1)
    X = np.c_[np.ones((len(X),1)) * 1, X] # bias = 1

    return X @ self.parameters
```
